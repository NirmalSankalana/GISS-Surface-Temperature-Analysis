```python
import pandas as pd
import numpy as np
import plotly.express as px
```

#### 1. Global-mean monthly, seasonal, and annual means:
1880-present, updated through most recent month

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/NirmalSankalana/GISS-Surface-Temperature-Analysis/main/GLB.Ts%2BdSST.csv', skiprows=[0])
df1.drop(['J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], axis=1, inplace=True);
df1.head()
```

#### 2. Northern Hemisphere-mean monthly, seasonal, and annual means
1880-present, updated through most recent month

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/NirmalSankalana/GISS-Surface-Temperature-Analysis/main/NH.Ts%2BdSST.csv', skiprows=[0])
df2.drop(['J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], axis=1, inplace=True);
df2.head()
```

#### 3. Southern Hemisphere-mean monthly, seasonal, and annual means
1880-present, updated through most recent month

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/NirmalSankalana/GISS-Surface-Temperature-Analysis/main/SH.Ts%2BdSST.csv', skiprows=[0])
df3.drop(['J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], axis=1, inplace=True);
df3.head()
```

```python
month =[]
temp_global = []
temp_nothern = []
temp_southern = []
for i in range(1880, 2022):
    for j in range(1, 13):
        temp_global.append(df1.iloc[i - 1880,j])
        temp_nothern.append(df2.iloc[i - 1880,j])
        temp_southern.append(df3.iloc[i - 1880,j])
        month.append(str(j) + '/' + str(i))

df = pd.DataFrame({'month':month, 'temp_global':temp_global, 'temp_nothern': temp_nothern, 'temp_southern': temp_southern})
df['month'] = pd.to_datetime(df['month'].str.strip(), format='%m/%Y')
df.to_csv('temperature.csv', index = False)
df.head()
```

```python
fig = px.line(df, x='month', y=["temp_global", "temp_nothern", "temp_southern"])
fig.show()
```
